## 1. Gráficos
### 1.1 Nuve de puntos

````python
import matplotlib.pyplot as plt
# from matplotlib.pyplot import scatter

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

## Nuve de puntos
plt.scatter(x, y)

# Agregar título
plt.title("Nuve de puntos")

# Los ejes
plt.xlabel("Eje x")
plt.ylabel("Eje y")

# Mostrar el gráfico
plt.show()
````

````python
# Graficar una nuve puntos para f(x) = x^2, x = [1, 2, 3, 4, 5]
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Gráficar una nuve de puntos
plt.scatter(x, y)
plt.show()
````

## 1.2 Lineas continuas

Se usa ```plt.plot```

````python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, color="red")
plt.scatter(x, y, color="green")
plt.show()
````
